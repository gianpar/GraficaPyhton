```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('insurance.csv')

sns.set_theme(style="whitegrid")
```

```python
df.info()
```

```python
df[['region', 'charges', 'sex', 'age']].head(10)
```

```python
# Grafico de conteo y Grafico de barras
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='region', palette='Set2')
plt.title('Cantidad de personas por región', fontsize=14)
plt.xlabel('Región (Ubicación Geografica)', fontsize=14)
plt.ylabel('Cantidad total de Registros', fontsize=14)
plt.show()

# Grafico de barras
sns.barplot(data=df, x='region', y='charges', palette='Set2')
plt.title('Analisis de costo promedio del seguro por región', fontsize=14)
plt.xlabel('Región (Ubicación Geografica)', fontsize=14)
plt.ylabel('Cposto promedio en dolares ($)', fontsize=14)
plt.show()

```

```python
#Grafica de dispersión

plt.figure(figsize=(8,5))

sns.scatterplot(data=df, x='age', y='charges', hue='sex', palette=['#F54927', '#27F5CC'])
plt.title('Relacion entre Edad y costo (Agrupado por fumador)', fontsize=14)
plt.xlabel('Edad', fontsize=14)
plt.ylabel('Costo del seguro($)', fontsize=14)
plt.show()

```

```python
# Grafico de lineas
plt.figure(figsize=(8,5))

sns.lineplot(data=df, x='age', y='charges', hue='sex', palette=['#F54927', '#27F5CC'])
plt.title('Tendencias de costos a lo largo de la edad', fontsize=1)
plt.xlabel('Edad', fontsize=14)
plt.ylabel('Costo del seguro($)', fontsize=14)
plt.show()

```

```python
# Grafico de tortas/pie
plt.figure(figsize=(8,5))

conteos_genero = df['sex'].value_counts()
plt.pie(conteos_genero, labels=['Hombres', 'Mujeres'], autopct='%1.1f%%', colors=['#F54927', '#27F5CC'])
plt.title('Proporción de hombres y mujeres')
plt.show()
```

```python
# Grafico histograma y densidad
plt.figure(figsize=(8,5))

sns.histplot(data=df, x='charges', kde=True, color='#27F5CC')
plt.title('Distribución de costos', fontsize=14)
plt.xlabel('Costo', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.show()

```

```python
#diagrama de caja
plt.figure(figsize=(8,5))

sns.boxplot(data=df, x='smoker', y='charges', palette='Set2')
plt.title('Distribución fumadores vs no fumadores', fontsize=1)
plt.show()
```

```python
#Grafica Striplot
plt.figure(figsize=(8,5))

sns.stripplot(data=df, x='children', y='charges', jitter=True, alpha=0.5, color='black')
sns.boxplot(data=df, x='children', y='charges', palette='Set2')
plt.title('Distribución fumadores vs no fumadores', fontsize=1)
plt.xlabel('Número de hijos', fontsize=14)
plt.ylabel('Costo del seguro ($)', fontsize=14)
plt.show()
```

```python
## Grafico de Regresión
plt.figure(figsize=(8,5))

sns.regplot(data=df, x='bmi', y='charges', scatter_kws={'color':'blue', 'alpha':0.3} ,line_kws={'color': 'red'}, )
plt.title('Regresión: indice de masa corporal vs costo', fontsize=14)
plt.xlabel('Indice de masa corporal', fontsize=14)
plt.ylabel('Costo del seguro ($)', fontsize=14)
plt.show()
```

```python
#mapa de calor
plt.figure(figsize=(8,5))

correlacion= df.select_dtypes(include='number').corr()
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.title('Mapa de calor: que tanto influye una variable en otra', fontsize=14)
plt.show()
```

```python
#Grafico red o multivariable

sns.jointplot(data=df, x='bmi', y='charges', kind='scatter', color='#27F5CC')
plt.suptitle('jointplot: dispersión + histogramas laterales', y=1.02)
plt.show()

sns.pairplot(df, hue='smoker', palette='Set2')
plt.suptitle('pairplot: grafico de pares', y=1.02)
plt.show()
```
